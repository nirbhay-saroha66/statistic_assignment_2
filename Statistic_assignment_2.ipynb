Q9) What is the Poisson distribution and where is it applied?

```python
"""
The Poisson distribution is a discrete probability distribution that models the
number of events occurring in a fixed interval of time or space, assuming that:
=>The events occur independently.
=>The average number of events per interval (λ) is constant.
=>Two events cannot occur at exactly the same instant.

1.Definition:
=>A random variable X follows a Poisson distribution if it represents the number
of occurrences of an event in a fixed interval (time, space, volume, etc.),
given a known average rate λ (mean number of events per interval).

2.Probability Mass Function (PMF):
=>The probability of exactly k occurrences in an interval is given by:

𝑃(𝑋=𝑘)= (e power -𝜆)x (𝜆 power 𝑘)/ 𝑘! , k = 0,1,2,...

3.Cumulative Distribution Function (CDF)
=>The probability of getting at most k occurrences:

𝐹(𝑘)=𝑃(𝑋≤𝑘)=∑(i=0 to k) (𝑒 power −𝜆) x (𝜆 power i)/𝑖!

4.Expected Value(Mean):
=> E[X] = λ

5.Variance:
=> Var(X) = λ

The Poisson distribution is widely used for modeling rare or random events occurring over time or space, such as:

=>Queuing Theory: Number of customers arriving at a service center per hour.
=>Telecommunications: Number of phone calls received per minute.
=>Traffic Flow: Number of cars passing a checkpoint in a given time.
=>Biology & Medicine: Number of mutations in a DNA strand per segment.
=>Economics: Number of defaults in a bank portfolio in a month.
=>Crime & Risk Analysis: Number of accidents at an intersection per week.
=>Astronomy: Number of meteors observed per hour.

"""
```

Q10) What is a continuous uniform distribution?

```python
"""
The continuous uniform distribution is a probability distribution where all
outcomes within a given range are equally likely.
It is often referred to as the rectangular distribution because its probability
density function (PDF) is a constant value, forming a flat, horizontal line when graphed.

1.Definition:
=>A random variable X follows a continuous uniform distribution if it is equally
likely to take any value in the interval [𝑎,𝑏], where:
a is the minimum value (lower bound).
b is the maximum value (upper bound).

2.Probability Density Function (PDF):
=>Since all values between a and b are equally likely, the probability density function is:
    f(x)= 1/(b-a) for a<=x<=b
          0, otherwise

3.Cumulative Distribution Function (CDF):
=>The cumulative probability of getting a value less than or equal to x is given by:
      F(x)= 0, x<a
            (x-a)/(b-a), a<=x<=b
            1, x>b
4.Expected Value(Mean):
=> E[X] = (a+b)/2

5.Variance:
=> Var(X) = (b-a)^2/12

The continuous uniform distribution is used in scenarios where outcomes are equally
likely over a range, including:

=>Random Number Generation: Generating random numbers in simulations.
=>Simulation & Monte Carlo Methods: Modeling uncertainty in physics and finance.
=>Time & Space Analysis: Modeling arrival times within a fixed period when no preference exists.
=>Manufacturing & Quality Control: When defects occur randomly within a measurement range.
"""
```

Q11) What are the characteristics of a normal distribution?

```python
"""
Characteristics of a Normal Distribution:
1.Bell-Shaped Curve:
=>The distribution is symmetrical around the mean μ.
=>The peak of the curve occurs at x=μ.

2.Mean, Median, and Mode are Equal:
=> In a normal distribution, μ (mean), median, and mode are the same.

3.Symmetry:
=>The curve is perfectly symmetrical about the mean.
=>The left half is a mirror image of the right half.

4.Empirical Rule (68-95-99.7 Rule):
=>About 68% of the data falls within 1 standard deviation (μ±σ).
=>About 95% falls within 2 standard deviations (μ±2σ).
=>About 99.7% falls within 3 standard deviations (μ±3σ).

5.Asymptotic Nature:
=>The tails of the normal distribution extend infinitely in both directions but
never touch the x-axis.

6.Unimodal:
=>The normal distribution has only one peak (single mode).

7.Area Under the Curve Equals 1:
=>The total probability for all possible values is 1.

8.Inflection Points:
=>The curve changes from concave down to concave up at x=μ±σ.
"""
```

Q13) What is the Central Limit Theorem (CLT), and why is it critical in statistics?

```python
"""
The Central Limit Theorem (CLT) is one of the most fundamental theorems in probability and statistics.
It states that:
Regardless of the original population distribution, the distribution of the
sample means (or sums) will approximate a normal distribution as the sample size
increases, provided the samples are independent and randomly selected.

This theorem is critical because it allows us to make inferences about a population
even when we don't know its original distribution.

CLT is important for the following reasons:
1.Enables Normal Approximation for Any Distribution:
=>Even if the population is skewed, uniform, or multimodal, the sample mean
distribution will still be approximately normal for large n.
=>This makes normal-based statistical methods widely applicable.

2.Foundation for Inferential Statistics:
=>Many statistical methods, including confidence intervals, hypothesis testing,
and Z-tests, rely on the normality assumption.
=>Even when the population distribution is unknown, we can estimate probabilities
using the normal distribution.

3.Justifies Use of the Normal Distribution in Sampling:
=>Since sample means tend to be normally distributed, we can use normal probability
calculations for inference.

4.Reduces the Need for Large Populations:
=>A well-chosen sample can represent the entire population, reducing time and cost.
"""
```

```python
"""
The Central Limit Theorem (CLT) and the normal distribution are closely connected
because CLT explains why the normal distribution appears so frequently in real-world
data and statistical analysis.

1.CLT Explains Why Sample Means Follow a Normal Distribution:
=>The normal distribution is a key outcome of the Central Limit Theorem.
=>CLT states that regardless of the shape of the original population distribution,
the distribution of sample means will approach a normal distribution as the sample size
increases.
=>This holds true even if the original data is skewed, uniform, or multimodal.

2.Why Does CLT Lead to a Normal Distribution?
=>The normal distribution results from summing many independent random variables.
=>Real-world data is often influenced by many small, independent factors, which
collectively form a normal-like shape.
=>This is why heights, test scores, and measurement errors often follow a normal
distribution.

3.How Large Should n Be?
=>If the original population is already normal, then even small samples are normally distributed.
=>If the original population is skewed, a sample size of n≥30 is typically enough.
=>If the population is highly skewed, a larger n (e.g., 50 or more) is needed.
"""
```

Q2) Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range for the discrete uniform distribution
low, high = 1, 6  # Example: Simulating a fair 6-sided die

# Generate random samples
samples = randint.rvs(low, high + 1, size=1000)

# Compute PMF values
x = np.arange(low, high + 1)
pmf_values = [randint.pmf(k, low, high + 1) for k in x]

# Plot the PMF
plt.figure(figsize=(8, 5))
plt.bar(x, pmf_values, width=0.6, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel("Values")
plt.ylabel("Probability")
plt.title("PMF of Discrete Uniform Distribution")
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Q3) Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

def bernoulli_pdf(p):
    """
    Plots the Probability Distribution Function (PDF) of a Bernoulli distribution.

    Parameters:
    p (float): Probability of success (0 ≤ p ≤ 1)
    """
    # Possible values for a Bernoulli random variable
    x = np.array([0, 1])

    # Compute PDF values
    pdf_values = bernoulli.pmf(x, p)

    # Plot the PDF
    plt.figure(figsize=(6, 4))
    plt.bar(x, pdf_values, width=0.4, color='skyblue', edgecolor='black', alpha=0.7)
    plt.xticks([0, 1])
    plt.xlabel("Outcome (X)")
    plt.ylabel("Probability P(X)")
    plt.title(f"Bernoulli Distribution (p = {p})")
    plt.ylim(0, 1)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    return dict(zip(x, pdf_values))  # Return PDF values as a dictionary

# Example usage
p = 0.7  # Probability of success
pdf_result = bernoulli_pdf(p)
print("Bernoulli PDF:", pdf_result)
```

Q4) Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters for the Binomial Distribution
n = 10   # Number of trials
p = 0.5  # Probability of success

# Generate random samples from the binomial distribution
samples = np.random.binomial(n, p, size=1000)  # Simulating 1000 trials

# Compute the theoretical binomial PMF
x = np.arange(0, n + 1)  # Possible values of X (0 to n)
pmf_values = binom.pmf(x, n, p)

# Plot histogram of simulated data
plt.figure(figsize=(8, 5))
plt.hist(samples, bins=np.arange(-0.5, n + 1.5, 1), density=True, alpha=0.6, color='skyblue', edgecolor='black', label="Simulated Data")

# Overlay the theoretical binomial PMF
plt.plot(x, pmf_values, 'ro-', markersize=8, label="Theoretical PMF")

# Labels and Title
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution Histogram (n={n}, p={p})")
plt.xticks(range(n + 1))
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

Q6) Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define range for the discrete uniform distribution
low, high = 1, 6  # Example: Rolling a fair 6-sided die

# Generate possible values
x = np.arange(low, high + 1)

# Compute CDF values
cdf_values = randint.cdf(x, low, high + 1)

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(x, cdf_values, where='mid', color='blue', label="Discrete Uniform CDF", linewidth=2)
plt.scatter(x, cdf_values, color='red', zorder=3)

# Labels and Title
plt.xlabel("X (Outcomes)")
plt.ylabel("Cumulative Probability")
plt.title(f"CDF of Discrete Uniform Distribution U({low}, {high})")
plt.xticks(x)
plt.yticks(np.arange(0, 1.1, 0.2))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, mean, std_dev):
    """
    Calculates the Z-score for a given data point.

    Parameters:
    - data_point: The value for which the Z-score is to be calculated
    - mean: Mean of the dataset
    - std_dev: Standard deviation of the dataset

    Returns:
    - Z-score
    """
    return (data_point - mean) / std_dev

def plot_standard_normal_with_z(z_score):
    """
    Plots the standard normal distribution and highlights the given Z-score.

    Parameters:
    - z_score: The calculated Z-score to highlight on the plot
    """
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, loc=0, scale=1)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, 'r-', lw=2, label="Standard Normal Distribution (N(0,1))")

    # Highlight the Z-score on the plot
    plt.axvline(z_score, color='blue', linestyle='--', linewidth=2, label=f"Z-score = {z_score:.2f}")

    # Shading area to the left of Z-score
    x_fill = np.linspace(-4, z_score, 500)
    plt.fill_between(x_fill, norm.pdf(x_fill), color='lightblue', alpha=0.5)

    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.title("Standard Normal Distribution with Z-score Highlighted")
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Example Usage
data_point = 75
mean = 50
std_dev = 15

# Calculate Z-score
z_score = calculate_z_score(data_point, mean, std_dev)
print(f"Z-score for data point {data_point}: {z_score:.2f}")

# Plot standard normal distribution with Z-score
plot_standard_normal_with_z(z_score)
```

Q19)  Implement hypothesis testing using Z-statistics for a sample dataset?

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Performs a one-sample Z-test for a given dataset.

    Parameters:
    - sample: List or numpy array of sample data
    - pop_mean: Population mean (μ)
    - pop_std: Population standard deviation (σ)
    - alpha: Significance level (default = 0.05)

    Returns:
    - Z-score, p-value, and test decision
    """
    # Sample statistics
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Compute Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    # Compute p-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Decision: Reject or fail to reject H₀
    decision = "Reject H₀" if p_value < alpha else "Fail to reject H₀"

    # Print results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z_score:.2f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Decision: {decision}")

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(loc=52, scale=15, size=30)  # Sample data with mean ~52
population_mean = 50   # Given population mean (μ)
population_std = 15     # Given population standard deviation (σ)

# Run the Z-test
z_test(sample_data, population_mean, population_std)
```

Q20)  Create a confidence interval for a dataset using Python and interpret the result?

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Computes the confidence interval for a dataset.

    Parameters:
    - data: List or numpy array of sample data
    - confidence: Confidence level (default = 95%)

    Returns:
    - Confidence Interval (Lower bound, Upper bound)
    """
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation
    sample_size = len(data)

    # Use t-distribution (if sample size is small) or normal distribution
    t_critical = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)

    # Margin of error
    margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

    # Confidence Interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    # Print results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Margin of Error: {margin_of_error:.2f}")
    print(f"{confidence*100:.0f}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    return lower_bound, upper_bound

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Generating a sample dataset

# Calculate 95% confidence interval
confidence_interval(sample_data)
```

Q27)  Implement Z-score normalization on a dataset and visualize its transformation

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def z_score_normalization(mean=50, std=15, size=1000):
    """
    Generates a dataset from a normal distribution, applies Z-score normalization,
    and visualizes the transformation using histograms and boxplots.

    Parameters:
    - mean: Mean of the original normal distribution
    - std: Standard deviation of the original distribution
    - size: Number of samples

    Returns:
    - Displays histograms and boxplots of the original and normalized data.
    """
    # Generate a normal dataset
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Z-score normalization
    mean_original = np.mean(data)
    std_original = np.std(data, ddof=1)
    normalized_data = (data - mean_original) / std_original  # Z-score formula

    # Create a DataFrame for visualization
    df = pd.DataFrame({'Original Data': data, 'Normalized Data': normalized_data})

    # Print statistics
    print(f"Original Data -> Mean: {mean_original:.2f}, Std Dev: {std_original:.2f}")
    print(f"Normalized Data -> Mean: {np.mean(normalized_data):.2f}, Std Dev: {np.std(normalized_data, ddof=1):.2f}")

    # Plot histograms before and after normalization
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    sns.histplot(data, bins=30, color='blue', alpha=0.6, kde=True)
    plt.axvline(mean_original, color='red', linestyle='--', label=f"Mean: {mean_original:.2f}")
    plt.title("Original Data Distribution")
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.legend()

    plt.subplot(1, 2, 2)
    sns.histplot(normalized_data, bins=30, color='green', alpha=0.6, kde=True)
    plt.axvline(0, color='red', linestyle='--', label="Mean: 0")
    plt.title("Z-Score Normalized Data Distribution")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.legend()

    plt.tight_layout()
    plt.show()

    # Boxplot to compare distributions
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df, palette=['blue', 'green'])
    plt.title("Boxplot of Original vs Normalized Data")
    plt.ylabel("Values")
    plt.xticks([0, 1], ["Original Data", "Normalized Data"])
    plt.grid(alpha=0.5)
    plt.show()

# Example Usage
z_score_normalization(mean=50, std=15, size=1000)
```

Q28) Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_skewness_kurtosis(mean=50, std=15, size=1000):
    """
    Generates a dataset from a normal distribution, calculates skewness and kurtosis,
    and visualizes the distribution using a histogram.

    Parameters:
    - mean: Mean of the normal distribution
    - std: Standard deviation of the normal distribution
    - size: Number of samples

    Returns:
    - Prints skewness and kurtosis values
    - Displays a histogram of the dataset
    """
    # Generate random data from a normal distribution
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness
    skewness = stats.skew(data)

    # Calculate kurtosis
    kurtosis = stats.kurtosis(data, fisher=True)  # Fisher=True returns excess kurtosis

    # Print results
    print(f"Skewness: {skewness:.4f} (Near 0 for normal distribution)")
    print(f"Kurtosis: {kurtosis:.4f} (Near 0 for normal distribution)")

    # Plot histogram with KDE (Kernel Density Estimate)
    plt.figure(figsize=(8, 5))
    sns.histplot(data, bins=30, kde=True, color='skyblue')
    plt.axvline(np.mean(data), color='red', linestyle='--', label=f"Mean: {mean}")
    plt.title("Histogram of Normally Distributed Data")
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(alpha=0.5)
    plt.show()

# Example Usage
calculate_skewness_kurtosis(mean=50, std=15, size=1000)
```
